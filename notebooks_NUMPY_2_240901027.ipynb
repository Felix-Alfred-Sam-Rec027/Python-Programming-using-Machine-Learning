```python
import numpy as np
arr = np.array([
    [1,2,3,4],
    [5,7,6,8],
    [9,0,11,12],
    [3,14,15,16]
])
temp=arr[[0,1,2,3],[3,2,1,0]].astype(float)
print("\nElements at indices (0,3),(1,2),(2,1),(3,0),""(3,0):\n",temp)
```

```python
import numpy as np
arr=np.array([[-1,2,0,4],
              [4,-0.5,6,0],
              [2.6,0,7,8],
              [3,-7,4,2.0]])
cond=arr>2
temp=arr[cond].astype(float)
print("\nElements greater than 2:\n",temp)
```

```python
import numpy as np
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
arr=np.concatenate((arr1,arr2))
print("\nOriginal arrays:\n",arr1,arr2) 
print("\nJoined array:\n",arr)
```

```python
arr=np.hstack((arr1,arr2))
print("\nHorizondal joining:\n",arr)
```

```python
arr=np.vstack((arr1,arr2))
print("\nVertical joining:\n",arr)
```

```python
arr=np.dstack((arr1,arr2))
print("\nDepth joining:\n",arr)
```

```python
arr=np.array([1,2,3,4,5,6])
newarr=np.array_split(arr,3)
print("\nOriginal Array:\n",arr)
print("\nSplitted array:\n",newarr)
```

```python
print("\nSplitted array in another form:\n")
print(newarr[0])
print(newarr[1])
print(newarr[2])
```
